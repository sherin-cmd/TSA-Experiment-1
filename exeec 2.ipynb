```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = "NLTK makes natural language processing fun. Python with NLTK is powerful. NLTK is fun too!"

tokens = word_tokenize(text)

fdist = FreqDist(tokens)

fdist.plot(10, title="Top 10 Word Frequencies")
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = "NLTK makes natural language processing fun. Python with NLTK is powerful. NLTK is fun too!"

tokens = word_tokenize(text)

fdist = FreqDist(tokens)

print("Vocabulary size:", len(fdist))  
print("\nWord Frequencies:")
for word, freq in fdist.items():
    print(word, ":", freq)

print("\nTop 5 most common words:")
print(fdist.most_common(5))

```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

text = "NLTK makes natural language processing fun. Python with NLTK is powerful. NLTK is fun too!"

tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("Original Tokens:")
print(tokens)
print("\nTokens after Removing Stopwords:")
print(filtered_tokens)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.text import Text

nltk.download('punkt', quiet=True)

text = """NLTK makes natural language processing fun. 
Python with NLTK is powerful. NLTK is fun too! 
Machine learning with Python is amazing. 
Natural language is very important in data science. 
Machine learning is the future. Natural language is key in AI."""

# tokenize
tokens = word_tokenize(text)

# wrap into NLTK Text
nltk_text = Text(tokens)

# find collocations
print("Collocations in the text:")
nltk_text.collocations()

```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import bigrams

nltk.download('punkt', quiet=True)

text = "NLTK makes natural language processing fun. Python with NLTK is powerful. NLTK is fun too!"
tokens = word_tokenize(text)
bi_grams = list(bigrams(tokens))

print("Bigrams in the text:")
print(bi_grams)
```
